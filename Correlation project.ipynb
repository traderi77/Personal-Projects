```python
import pandas as pd
import numpy as np
import seaborn as sns


import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

df = pd.read_csv(r'C:\Users\lukas\Desktop\movies.csv')
```

```python
df.drop_duplicates()


#checking for outliers
df.boxplot(column=['gross'])
```

```python
#for better visualization, data is now ordered
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
#can't print due to github restrictions
```

```python
sns.regplot(x="gross", y="budget", data=df)
```

```python
corr_matrix = df.corr()

sns.heatmap(corr_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
df.apply(lambda x: x.factorize()[0]).corr(method='pearson')
#random numeric value assigned
```

```python
corr_matrix = df.apply(lambda x: x.factorize()[0]).corr(method='pearson')

sns.heatmap(corr_matrix, annot = True)

plt.title("Correlation matrix for Movies")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
corr_matrix = df.apply(lambda x: x.factorize()[0]).corr()

corr_pairs = corr_matrix.unstack()

print(corr_pairs)
```

```python
sortet_pairs = corr_pairs.sort_values(kind="quicksort")

print(sortet_pairs)
```

```python
act_correlation = sortet_pairs[abs(sortet_pairs) > 0.5]

print(act_correlation)
```
